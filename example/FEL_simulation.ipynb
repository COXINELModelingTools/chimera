# Importing the code, numpy and graphics utilities

```python
from chimera.moduls.species import Specie
from chimera.moduls.solvers import Solver
from chimera.moduls.chimera_main import ChimeraRun
from chimera.moduls.diagnostics import Diagnostics
import chimera.moduls.fimera as chimera

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import clear_output
```

# Defining the simulation parameters

## Undulator

```python
K0 = 1.95
Periods = 300
lam0 = 2.8
```

## Electron beam

```python
g0 = 200/0.511
lbx  = 80e-4/lam0
lbr  = 90e-4/lam0
dens0 = 0.8e14/1.1e21*(lam0*1e4)**2
```

## Simulation grid

```python
Lgx  = 200e-4  /lam0
Rg   = 2000e-4  /lam0
Rg_cut = 800e-4/lam0
Nx,Nr = 120,80
```

## Preparing input

```python
dt = 1./40
Steps2Do = int(Periods/dt)+1

BoxGrid     = (-0.5*Lgx, 0.5*Lgx, Rg, Lgx/Nx, Rg/Nr)
BoxGridBeam = (-0.5*Lgx, 0.5*Lgx, Rg, Lgx/Nx, lbr/Nr)
Domain = (-0.5*lbx, 0.5*lbx, 0.0, lbr)

k_res = 2*g0**2/(1+K0**2/2)
gg = g0/(1.+K0**2/2)**.5
vb = (1.-gg**-2)**0.5

micro_scl = lam0*1e4
nano_scl = lam0*1e7
```

## Arguments for the simulation components

```python
beam_in = {
    'Charge':-1.,'Mass':1.,'Density':dens0, 'MomentaMeans':(g0,0.,0.), 'MomentaSpreads':(1e-4*g0,0.008,0.008),
    'Grid':BoxGridBeam, 'TimeStep':dt,'RandCell':50,'Xchunked':(4,6),
    'Devices':([chimera.undul_analytic,np.array([K0, 1., 1., 1000.])],)
}

solver_in = {
    'Grid':BoxGrid, 'TimeStep':dt, 'MaxAzimuthMode':0, 'KxShift':k_res,'CoPropagative':vb,
    'Rcut':Rg_cut ,'Xchunked':(4,6)
}

seed_in = {
    'a0':0.107,'k0':k_res,'x0':-30e-4/lam0,'x_foc':70.0/lam0,'Lx':15e-4/lam0,'LR':180e-4/lam0
}

MovingFrame = {
    'TimeStep':dt,'Steps':1,'Velocity':vb,'Features':('Staged','NoSorting')
}
```

# Constructing the simulation components

```python
beam = Specie(beam_in)
beam.add_particles(beam.gen_parts(Domain=Domain))
beam.denoise((k_res,))
solver = Solver(solver_in)
solver.add_gauss_beam(seed_in)

chimera_in = {'Solvers':(solver,),'Particles':(beam,),'MovingFrames':(MovingFrame,)}
Chimera = ChimeraRun(chimera_in)

Diags = Diagnostics(Chimera,(),out_folder=None)
```

# Running the simulation

```python
Chimera.make_halfstep()

diag_step = int(1./dt)
nrg = Diags.nrg_out({'Features':('Return',)})[0]
pwr = Diags.pwr_out({'Features':('Return',)})[0]

diag_nrg = np.zeros((int(Steps2Do/diag_step)+1,nrg.shape[0]))
diag_pwr = np.zeros((int(Steps2Do/diag_step)+1,pwr.shape[0]))

diag_nrg[0] = nrg
diag_pwr[0] = pwr

for i in xrange(1,Steps2Do):
    Chimera.make_step(i)
    if np.mod(i,diag_step)==0:
        clear_output(wait=True)
        print str(i)+'/'+str(Steps2Do-1)
        diag_nrg[int(i/diag_step)] = Diags.nrg_out({'Features':('Return',)})[0]
        diag_pwr[int(i/diag_step)] = Diags.pwr_out({'Features':('Return',)})[0]
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(20,7))
ax1.semilogy(diag_nrg.sum(-1))
ax2.plot(beam.Args['Xgrid'], diag_pwr[-1])
ax2.plot(beam.Args['Xgrid'], diag_pwr[0])
```
